# Observation:- Since the value of Outlet_Location_type is greater than 5.So it has multicolinearity problem.

```python
## Removing outlet_Location_type
df.drop('Outlet_Location_Type',axis=1,inplace=True)
df2.drop('Outlet_Location_Type',axis=1,inplace=True)
```

```python
X=df.drop('Item_Outlet_Sales',axis=1)
y=df.Item_Outlet_Sales
```

```python
x_scaled=scaler.fit_transform(X)
```

```python
x_scaled.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=41)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_train)
```

```python
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_score(y_train,ypred)
```

```python
y_pred=lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val=cross_val_score(lr, x_test,y_test,cv=5)
cross_val.mean()
```

```python
print('MSE  -',mean_squared_error(y_test,y_pred))
print('MAE  -',mean_absolute_error(y_test,y_pred))
```

```python
from xgboost import XGBRegressor
```

```python
xgb=XGBRegressor()
```

```python
xgb.fit(x_train,y_train)
```

```python
ypred=xgb.predict(x_train)
```

```python
r2_score(y_train,ypred)
```

```python
y_pred=xgb.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val=cross_val_score(xgb, x_test,y_test,cv=5)
cross_val.mean()
```

```python
print('MSE  -',mean_squared_error(y_test,y_pred))
print('MAE  -',mean_absolute_error(y_test,y_pred))
```
